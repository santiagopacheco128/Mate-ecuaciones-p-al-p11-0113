#### Problema 1
 $1x + 8 = 15$
 Identificar términos: Tenemos una incógnita $x$ sumada a $8$.Transposición: El $+8$ pasa al otro lado de la igualdad con el signo opuesto (restando).$1x = 15 - 8$Resolver operación:$1x = 7$Resultado: $x = 7$


```python
# Problema 1: Hallar x tal que 1x + 8 = 15
resultado_total = int(input("Ingrese el resultado total : "))
numero_sumado = int(input("Ingrese el número que suma : "))
coeficiente_x = 1

x = (resultado_total - numero_sumado) / coeficiente_x

print(f"El valor de x es: {x}")
```

#### Problema 3
 $3x + 4x = 21$
 Agrupar términos semejantes: Sumamos los coeficientes de $x$ ($3 + 4$).$7x = 21$Despejar la incógnita: El $7$ que está multiplicando a $x$ pasa al otro lado dividiendo.$x = 21 / 7$Resultado: $x = 3$

```python
# Problema 3: Hallar x tal que 3x + 4x = 21
resultado_total = int(input("Ingrese el resultado total : "))
coef_1 = int(input("Ingrese el primer coeficiente de x : "))
coef_2 = int(input("Ingrese el segundo coeficiente de x : "))

suma_coeficientes = coef_1 + coef_2
x = resultado_total / suma_coeficientes

print(f"El valor de x es: {x}")
```

#### Problema 5
$6a - 2a + a = 20$
Agrupar términos semejantes: Operamos los coeficientes de $a$ ($6 - 2 + 1$).$5a = 20$Despejar la incógnita: El $5$ pasa dividiendo al $20$.$a = 20 / 5$Resultado: $a = 4$

```python
# Problema 5: Hallar a tal que 6a - 2a + a = 20
resultado_total = int(input("Ingrese el resultado total : "))
# Coeficientes: 6, -2, 1
c1, c2, c3 = 6, -2, 1

suma_coeficientes = c1 + c2 + c3
a = resultado_total / suma_coeficientes

print(f"El valor de a es: {a}")
```

#### Problema 7
$4x + 3 = 2x + 11$
Organizar la ecuación: Pasamos las $x$ al lado izquierdo y los números al derecho (cambiando sus signos).$4x - 2x = 11 - 3$Reducir términos:$2x = 8$Despejar:$x = 8 / 2$Resultado: $x = 4$

```python
# Problema 7: Hallar x tal que 4x + 3 = 2x + 11
print("Ecuación: 4x + 3 = 2x + 11")
valor_derecha = int(input("Ingrese el número constante de la derecha : "))
valor_izquierda = int(input("Ingrese el número constante de la izquierda : "))
coef_izq = int(input("Ingrese el coeficiente x de la izquierda : "))
coef_der = int(input("Ingrese el coeficiente x de la derecha : "))

# Despejamos: (4x - 2x) = 11 - 3
# 2x = 8 -> x = 8 / 2
x = (valor_derecha - valor_izquierda) / (coef_izq - coef_der)

print(f"El valor de x es: {x}")
```

#### Problema 9
$9m + 4 = 5m + 20$
Organizar la ecuación: Movemos términos para agrupar variables y constantes.$9m - 5m = 20 - 4$Reducir términos:$4m = 16$Despejar:$m = 16 / 4$Resultado: $m = 4$

```python
# Problema 9: Hallar m tal que 9m + 4 = 5m + 20
print("Ecuación: 9m + 4 = 5m + 20")
valor_derecha = int(input("Ingrese el número constante de la derecha : "))
valor_izquierda = int(input("Ingrese el número constante de la izquierda : "))
coef_izq = int(input("Ingrese el coeficiente m de la izquierda : "))
coef_der = int(input("Ingrese el coeficiente m de la derecha : "))

# Despejamos: (9m - 5m) = 20 - 4
m = (valor_derecha - valor_izquierda) / (coef_izq - coef_der)

print(f"El valor de m es: {m}")
```

#### Problema 11
 $6a + 5 = 2a + 21$
 Organizar la ecuación: Movemos las $a$ a la izquierda y los números a la derecha.$6a - 2a = 21 - 5$Reducir términos:$4a = 16$Despejar:$a = 16 / 4$Resultado: $a = 4$

```python
# Problema 11: Hallar a tal que 6a + 5 = 2a + 21
print("Ecuación: 6a + 5 = 2a + 21")
valor_derecha = int(input("Ingrese el número constante de la derecha : "))
valor_izquierda = int(input("Ingrese el número constante de la izquierda : "))
coef_izq = int(input("Ingrese el coeficiente a de la izquierda : "))
coef_der = int(input("Ingrese el coeficiente a de la derecha : "))

# Despejamos: (6a - 2a) = 21 - 5
a = (valor_derecha - valor_izquierda) / (coef_izq - coef_der)

print(f"El valor de a es: {a}")
```
